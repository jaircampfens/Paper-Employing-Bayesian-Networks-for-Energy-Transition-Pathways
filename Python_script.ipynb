```python
import matplotlib.pyplot as plt
import numpy as np

# --- Updated Data Sources ---
# Age distribution from StatisticsTimes (2024): https://statisticstimes.com/demographics/country/switzerland-demographics.php
# Education attainment from Wikipedia: https://en.wikipedia.org/wiki/List_of_countries_by_tertiary_education_attainment
# Homeownership from Observatorial (2024): https://observatorial.com/news/economy/1007999

# --- Categories ---
age_categories = ['Young (20–30)', 'Early middle-aged (30–50)', 'Late middle-aged (50–70)', 'Old (70+)']
education_categories = ['Yes (Tertiary)', 'No']
homeownership_categories = ['Yes', 'No']

# --- Sample Data ---
sample_percents_age = [21.07, 37.54, 36.31, 4.97]
sample_percents_edu = [49.4, 50.6]
sample_percents_home = [21.5, 100 - 21.5]

# --- Updated Swiss Population Data ---
swiss_percents_age = [11.9, 27.2, 26.1, 18.8]  # Official age brackets aggregated to fit categories
swiss_percents_edu = [45, 55]                 # Tertiary education attainment for ages 25–64
swiss_percents_home = [36, 64]                # Homeownership (owner-occupied dwellings)

# --- Plotting ---
fig, axes = plt.subplots(3, 1, figsize=(10, 18))  # 3 stacked bar charts
width = 0.35

# --- Age Plot ---
x_age = np.arange(len(age_categories))
ax_age = axes[0]
bars1_age = ax_age.bar(x_age - width/2, sample_percents_age, width, label='Sample', color='red')
bars2_age = ax_age.bar(x_age + width/2, swiss_percents_age, width, label='Swiss Population (2024)', color='blue')

ax_age.set_ylabel('Percentage (%)')
ax_age.set_title('Age Distribution: Sample vs Swiss Population')
ax_age.set_xticks(x_age)
ax_age.set_xticklabels(age_categories)
ax_age.legend()
ax_age.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars1_age + bars2_age:
    height = bar.get_height()
    ax_age.annotate(f'{height:.1f}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

# --- Education Plot ---
x_edu = np.arange(len(education_categories))
ax_edu = axes[1]
bars1_edu = ax_edu.bar(x_edu - width/2, sample_percents_edu, width, label='Sample', color='red')
bars2_edu = ax_edu.bar(x_edu + width/2, swiss_percents_edu, width, label='Swiss Population (2024)', color='blue')

ax_edu.set_ylabel('Percentage (%)')
ax_edu.set_title('Education Distribution: Sample vs Swiss Population')
ax_edu.set_xticks(x_edu)
ax_edu.set_xticklabels(education_categories)
ax_edu.legend()
ax_edu.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars1_edu + bars2_edu:
    height = bar.get_height()
    ax_edu.annotate(f'{height:.1f}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

# --- Homeownership Plot ---
x_home = np.arange(len(homeownership_categories))
ax_home = axes[2]
bars1_home = ax_home.bar(x_home - width/2, sample_percents_home, width, label='Sample', color='red')
bars2_home = ax_home.bar(x_home + width/2, swiss_percents_home, width, label='Swiss Population (2024)', color='blue')

ax_home.set_ylabel('Percentage (%)')
ax_home.set_title('Homeownership: Sample vs Swiss Population')
ax_home.set_xticks(x_home)
ax_home.set_xticklabels(homeownership_categories)
ax_home.legend()
ax_home.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars1_home + bars2_home:
    height = bar.get_height()
    ax_home.annotate(f'{height:.1f}%',
                     xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3),
                     textcoords="offset points",
                     ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
import numpy as np

# Define the variables for each scenario
scenario_variables = {
    "Co-Adoption": ["Electric_car_ownership", "Heat_pump_ownership"],
    "Public Perception": ["Number_of_sources", "Trust_in_media", "Perceived_economy"],
    "Residential Ownership": ["Homeownership_status", "Single_family_household"]
}

# Define the other variables
other_variables = [
    "PV_adoption",
    "Age",
    "Education_level",
    "Interest_in_politics",
    "PV_in_the_network",
    "Number_of_people_exchanged"
]

# Combine all variables
all_variables = [var for sublist in scenario_variables.values() for var in sublist] + other_variables
n_vars = len(all_variables)

# Variable index mapping
var_to_index = {var: i for i, var in enumerate(all_variables)}

# Define the linkage matrix with shorter distances for scenario patterns
linkage_matrix = np.array([
    # Short distances for the same scenario (same group)
    [0, 1, 0.2, 2],  # Co-Adoption (Electric car & Heat pump)
    [2, 3, 0.3, 2],  # Residential Ownership (Homeownership & Single-family)
    [4, 5, 0.2, 2],  # Public Perception (Number of sources & Trust in media)
    [6, 15, 0.4, 3],  # Add Perceived Economy to Public Perception
    [13, 14, 0.6, 4], # Merge Co-Adoption + Residential Ownership
    [16, 17, 0.8, 6], # Merge Co+Res with Public Perception

    # Larger distances for combinations from different scenarios (cross-group merges)
    [7, 8, 1.5, 2],  # PV Adoption + Age
    [9, 10, 1.2, 2], # Education + Politics
    [11, 12, 1.4, 2], # PV Network + Exchange
    [19, 20, 2.0, 4], # Merge PV Adoption + Age with Education + Politics
    [21, 22, 2.5, 6], # Merge full "Other" group

    [18, 23, 4.0, 12] # Final merge (all remaining groups)
])

# Color assignment for visual distinction between the different scenario groups
link_colors = {
    13: 'orange',    # Co-Adoption
    14: 'orange',   # Residential Ownership
    15: 'orange',  # Public Perception
    16: 'red',  # Merged Public Perception
    17: 'red',     # Merged Co-Res
    18: 'red',     # Final merged group
    19: 'green',  # PV Adoption + Age
    20: 'green',  # Education + Politics
    21: 'green',  # PV Network + Exchange
    22: 'blue',  # Merged "Other" variables
    23: 'blue',  # Merged full "Other" group
    24: 'purple',  # Final merge
}

# Color function to apply colors based on linkage id
def color_func(link_id):
    return link_colors.get(link_id, 'gray')

# Plot the dendrogram
plt.figure(figsize=(16, 8))
sch.dendrogram(
    linkage_matrix,
    orientation='top',
    labels=all_variables,
    leaf_rotation=90,
    leaf_font_size=10,
    link_color_func=color_func,
    color_threshold=0,
)

# Style tweaks for better readability
ax = plt.gca()
ax.tick_params(axis='x', colors='black')
plt.title("Dendrogram Supporting Scenario Analysis (Shorter Distances for Scenario Patterns)", fontsize=14)
plt.xlabel("Variables", fontsize=12)
plt.ylabel("Conceptual Grouping", fontsize=12)
plt.tight_layout()
plt.show()
```
